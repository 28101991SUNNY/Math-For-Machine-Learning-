#### Given equation of a hyperplane w1x1 + w2x2 + w3x3 + w0 = 0. How many features are there in this equation?


    Suppose you have a plane in three-dimensional space defined by a (non-zero) vector n = (a, b, c) normal to it.
    That means that the set of points v = (x, y, z) in space that lies on the plane are exactly those for which n*v = 0.
    A hyperplane is the n-1 dimensional solution set of one linear equation : a1x1 + a2x2 + …+ anxn = 0.

    Here the number of variables and features in the equation is 3.

```python
import math

def dotproduct(v1,v2):
    return sum([x*y for x,y in zip(v1,v2)])
```

```python
def magnitude(v):
    return math.sqrt(dotproduct(v,v))

```

```python
def angleBetweenTwoVectors(v1,v2):
    return math.acos( (dotproduct(v1,v2)) / (magnitude(v1)*magnitude(v2)) )
```

#### Slope and Intercept

#### The general equation of a line is written as: ax + by + c = 0 which can also be represented in terms of slope (m) and intercept (c) as L : y = mx + c  What is the slope and intercept of ax + by + c = 0?


m = -a/b ; c = -c/b

$ax + by + c = 0$ is general equation and $y = mx + c$
can be represented in terms of slope and intercept
Step 1: $-by = ax + c$

Step 2: $y = -\frac{a}{b}x$ + $-\frac{c}{b}$ [dividing both sides by -b]


Comparing this with the form $y = mx + c$,


we will get $m = -\frac{a}{b}$ and $c = -\frac{c}{b}$
