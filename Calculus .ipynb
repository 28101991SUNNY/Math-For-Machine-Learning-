As $Q$ becomes infinitely close to $P$:

Using the delta method, we've derived the definition of differentiation from first principles. The derivative of $y$ (denoted $dy$) with respect to $x$ (denoted $dx$) can be represented as: 
$$\frac{dy}{dx} = \lim_{\Delta x \to 0} \frac{\Delta y}{\Delta x}$$



Expanding $\Delta y$ out to $y_2 - y_1$: 
$$\frac{dy}{dx} = \lim_{\Delta x \to 0} \frac{y_2 - y_1}{\Delta x}$$



Finally, replacing $y_1$ with $f(x)$ and replacing $y_2$ with $f(x + \Delta x)$, we obtain a common representation of differentiation:
$$\frac{dy}{dx} = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x}$$

```python
def func(x):
    y = x**2 + 2*x + 2
    return y
```

```python
def diff_Fn(func,x,delta_x):
    return (func(x + delta_x) - func(x)) / delta_x
```

```python
delta_x = np.array([1,0.1,0.01,0.001,0.0001,0.00001,0.000001])
```

```python
for delta in delta_x:
    print(diff_Fn(func, 2, delta))
```

```python
diff_Fn = np.vectorize(diff_Fn)
```

```python
diff_Fn(func,2,np.array([1,0.1,0.01,0.001,0.0001,0.00001,0.000001]))
```

```python

def regression_plot(x, y, m, b):
    
    fig, ax = plt.subplots()

    ax.scatter(x, y)
    
    x_min, x_max = ax.get_xlim()
    y_min = regression(x_min, m, b).detach().item()
    y_max = regression(x_max, m, b).detach().item()
    
    ax.set_xlim([x_min, x_max])
    _ = ax.plot([x_min, x_max], [y_min, y_max])
```

### Partial Derivatives of Multivariate Functions
Define a function $f(x, y)$ for $z = x^2 - y^2$: 

```python
def f(x,y):
    return x**2 - y**2
```

```python
xs = np.linspace(-3,3,1000)
```

```python
z_wrt_x = f(xs,0)
```

```python
fig,ax = plt.subplots()
plt.axvline(x = 0, color = "lightgrey")
plt.axhline(y = 0, color = "lightgrey")
plt.xlabel("x")
plt.ylabel("z",rotation = 0)
plt.plot(xs,z_wrt_x)
```

    dz/dx = 2x

```python
def delzbydelx(x,y):
    return 2*x
```

```python
x_samples = [-3,-2,-1,0,1,2,3]
```

```python
colors = ["grey","red","orange","green","blue","purple","lightblue"]
```

```python
def point_and_tangent_wrt_x(xs,x,y,f,fderivative,col):
    z = f(x,y)
    plt.scatter(x,z,c = col,zorder = 3)
    
    tangent_m = fderivative(x,y) # slope  
    tangent_b = z - tangent_m * x   # calculating b = z-mx
    tangent_line = tangent_m * xs + tangent_b
    
    plt.plot(xs,tangent_line,c =col,
            linestyle = "dashed",
            linewidth = 0.7,
            zorder = 3)
```

```python
fig,ax = plt.subplots()
plt.axvline(x = 0, color = "lightgrey")
plt.axhline(y = 0, color = "lightgrey")

for i,x in enumerate(x_samples):
    point_and_tangent_wrt_x(xs,x,0,f,delzbydelx,colors[i])
    
plt.ylim(-1,9)
plt.xlabel("x")
plt.ylabel("z")
ax.plot(xs,z_wrt_x)
```

Thereby visually demonstrating $\frac{\partial z}{\partial x} = 2x$. 

That is, the slope of $z$ along the $x$ axis is *twice* the $x$ value. 

```python
# now z with respect to y by varying y

ys = np.linspace(-3,3,1000)
```

```python
z_wrt_y = f(2,ys)
```

```python
fig,ax = plt.subplots()
plt.axvline(x = 0, color = "lightgrey")
plt.axhline(y = 0, color = "lightgrey")
plt.xlabel("y")
plt.ylabel("z",rotation = 0)
plt.plot(ys,z_wrt_y)
```

As in the slides, the partial derivative of $z$ w.r.t $y$ happens to be independent of $x$ (just as we observed $x$ is independent of $y$ above), so while $z$ varies as a function of both $x$ and $y$, the slope of $z$ w.r.t $y$ is the same no matter what $x$ is: $$ \frac{\partial z}{\partial y} = -2y $$

So for example, holding $x$ constant at 2 instead of 0 increases $z$, but has no impact whatsoever on the slope of $z$ w.r.t. $y$: 

```python
def delz_by_dely(x, y):
    return -2*y
```

```python
y_samples = [-3,-2, -1, 0, 1, 2,3]
```

```python
def point_and_tangent_wrt_y(ys,x,y,f,fderivative,col):
    z = f(x,y)
    plt.scatter(y,z,c = col,zorder = 3)
    
    tangent_m = fderivative(x,y) # slope  
    tangent_b = z - tangent_m * y   # calculating b = z-mx
    tangent_line = tangent_m * ys + tangent_b
    
    plt.plot(ys,tangent_line,c =col,
            linestyle = "dashed",
            linewidth = 0.7,
            zorder = 3)
```

```python
fig,ax = plt.subplots()
plt.axvline(x = 0, color = "lightgrey")
plt.axhline(y = 0, color = "lightgrey")

for i,y in enumerate(y_samples):
    point_and_tangent_wrt_y(ys,2,y,f,delz_by_dely,colors[i])
    
plt.ylim(-6,6)
plt.xlabel("y")
plt.ylabel("z")
ax.plot(xs,z_wrt_y)
```

Thereby visually demonstrating $\frac{\partial z}{\partial y} = -2y$. 

That is, the slope of $z$ along the $y$ axis is *twice* the $y$ value *and inverted*, resulting in the parabolic curve opening downward.
