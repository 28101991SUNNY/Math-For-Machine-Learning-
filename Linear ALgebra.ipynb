### how matrices Transform space

```python
TransformationMatrix = np.array([[2,3],
                                [10,1]])
Vector = np.array([3,2])
```

```python
np.dot(TransformationMatrix,Vector)
```

```python
np.dot(TransformationMatrix,( Vector[0] * np.array([1,0]) + Vector[1] * np.array([0,1])))
```

```python
# matrix the doesnt changes anything
```

```python
TM = np.array([[1,0],
              [0,1]])    #Identity Matrix

Vector = np.array([3,2])

TM @  Vector
```

    1 0                      -1 0
    0 1  transformed into     0 2  

```python
TM = np.array([[-1,0],
               [0 ,2]])   

Vector = np.array([3,2])

TM @  Vector
```

```python
TM = np.array([[-1,0],  # flip the matrix 
               [0 ,-1]])   

Vector = np.array([3,2])

TM @  Vector
```

```python
TM = np.array([[0,1],  # mirror at 45 degree
               [1,0]])   

Vector = np.array([3,2])

TM @  Vector
```

```python
TM = np.array([[0,-1],  #  flip and mirror at 45 degree
               [-1,0]])   

Vector = np.array([3,2])

TM @  Vector
```

```python
TM = np.array([[-1,0],  #  vertical mirror
               [0,1]])   

Vector = np.array([3,2])

TM @  Vector
```

```python
# sheer transformation
```

```python
TM = np.array([[1,1],  
               [0,1]])   

Vector = np.array([3,2])

TM @  Vector
```

```python
# rotation  90 degree left
TM = np.array([[0,1],  
               [-1,0]])   

Vector = np.array([3,2])

TM @  Vector
```
