### Perceptron

- In machine learning, the perceptron (or McCulloch-Pitts neuron) is an algorithm for supervised learning of binary classifiers.
- A binary classifier is a function which can decide whether or not an input, represented by a vector of numbers, belongs to some specific class.
- It is a type of linear classifier, i.e. a classification algorithm that makes its predictions based on a linear predictor function combining a set of weights with the feature vector.

```python
w = np.random.normal(size = (2,))
w
```

```python
w0 = np.random.normal()
w0
```

```python
def show(features ,label,w,w0):  # x , y are data points
                    # w is weight (hyperplane wTx+ w0 = 0)
    
    
    slope = -w[0]/w[1]    # -a/b
    intercept = -w0/w[1]  # -c/b
    features_hyperplane = np.arange(10)
    label_plane = slope * features_hyperplane + intercept  # y = mx + c
    
    plt.plot(features_hyperplane,label_plane)
    plt.plot(features[:,0][label == 1], features[:,1][label == 1], "g^")
    plt.plot(features[:,0][label == -1], features[:,1][label == -1], "bs")
#     plt.xlim(-2,12)
#     plt.ylim(-2,12)
    plt.show()
```

```python
show(features,label,w,w0)
```

```python
def algorithm(features,label):

    # taking radom w and w0
    w = np.random.normal(size = (2,))
    w0 = np.random.normal()

    
    for iteration in range(5000):
        miss_classified_points = 0
        for p in range(len(features)):  # go over all data points
            halfspace = np.sign(np.dot(w,features[p]) + w0) # half space (sign)
            if halfspace != np.sign(label[p]):
                miss_classified_points += 1 # count missclasified points
                w = w + (features[p] * np.sign(label[p]))
                w0 = w0  + np.sign(label[p])
                break
                
        if miss_classified_points == 0 : 
            show(features,label,w,w0)
            print("iterations : ",iteration)
            print("Algorithm converged. Good Decision boundary is achieved.")
            break
        if iteration % 20 == 0:
            show(features,label,w,w0)
            print("Accuracy :",1- round(miss_classified_points/len(features),2))
    
    return w, w0
                
            
```

```python
algorithm(features,label)
```
