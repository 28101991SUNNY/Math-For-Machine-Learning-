#### The incomplete code snippet given below will be used to find out the angle between two vectors. 



    Create a function dotProduct to calculate the dot product of v1 and v2
    Create a function magnitude to calculate the mangnitude of a vector by
    calculating the square root of its dot product with itself.
    Create a function to calculate the angle between the two vectors,
    In order to perform their dot product, first check if their lengths are equal
    Calculate the angle between them by calculating the arc cosine of their dot product divided by the product of their mangnitude.


```python
import math

def dotproduct(v1,v2):
    return sum([x*y for x,y in zip(v1,v2)])
```

```python
def magnitude(v):
    return math.sqrt(dotproduct(v,v))

```

```python
def angleBetweenTwoVectors(v1,v2):
    return math.acos( (dotproduct(v1,v2)) / (magnitude(v1)*magnitude(v2)) )
```

### Straight line or not

### You are given a list of (X, Y) coordinates, check whether the coordinates lie on a straight line or not. If they lie on a straight line return the slope and intercept else return -1.

```python
def straight_line_or_not(coordinates):
    slopes = []
    for point in range(1,len(coordinates)):

        p1 = coordinates[point-1]
        p2 = coordinates[point]
    #     print(p1,p2)
        m = (p2[1]-p1[1])/(p2[0]-p1[0])
        slopes.append(m)

    ans = -1

    if len(set(slopes)) == 1:
        c = p1[1] - (m * p1[0])  #(y = mx + c : c = y-mx)
        ans = round(m,1),round(c,1)
    return (ans)
    
```

```python
coordinates = [(1.0, 5.0), (-3.0, -3.0), (2.5, 8.0)]
straight_line_or_not(coordinates)
```

    We define a function to check if given input (coordinates), lie on the same line

        If they do, we calculate the slope and intercept, else we return -1
        Store coordinates in an object named “ls”
        initlialize i as 0, which we will use to index our input list
        Define an empty list m, which we will use to store slope of our points
        We define a loop to run until i is 1 less than the coordinates length
        We calculate the slope for the for the consecutive points in our list,
        by using the formula :
        slope = (y2-y1)/(x2-x1)
        We check if the points are one same line i.e. slope = 1
        Then Compute their intercept using formula :
        intercept = y1 - slope*x1
        Round off the slope and intercept to 1 decimal
        If the points do not lie on the same line,
        i.e. slope is not equal to 1, we return -1

    def straight_line_or_not(coordinates):
        ls= coordinates
        i=0
        m=[]
        while i<len(ls)-1:
            slope=(ls[i+1][1]-ls[i][1])/(ls[i+1][0]-ls[i][0])
            m.append(slope)
            i=i+1
        if len(set(m))==1:
            c= ls[0][1]-m[0]*ls[0][0]
            return (round(m[0],1),round(c,1))
        else:
            return -1

###  Continuity & Differentiability
#### Given below are some statements related to continuity and differentiability
I) Every differentiable function is continuous

II) Every continuous function is differentiable

#### Which of the above statements are correct?

    I

    In order for a function to be differentiable within some given range it must be continuous,
    but continuity alone cannot be the judge for it to be differentiable.
    Take for example f(x) = |x|, this function is continuous but it is not differentiable at x = 0.
    So, a continuous function need not be differentiable, but a differentiable function is always continuous.


### Sigmoid function derivative



 f(x) x (1 - f(x))

Explanation:

Here $f(x) = \frac{1}{1 + e^{-x}}$

$\frac{\mathrm{d} }{\mathrm{d} x}f(x) = \frac{e^{-x}}{{(1 + e^{-x})}^{2}}$ [using quotient rule]

= $\frac{1}{1 + e^{-x}}$ x $\frac{e^{-x}}{1 + e^{-x}}$

= $\frac{1}{1 + e^{-x}}$ x ($1 - \frac{1}{1 + e^{-x}})$

= $f(x)$ x $(1 - f(x))$
